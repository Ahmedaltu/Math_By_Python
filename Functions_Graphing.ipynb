## Problem 9

Write a code that compares $sin(x)$ and its 1st, 3rd, 5th, and 7th order Taylor polynomial approximations in the same graph in range $x \in [-\pi, +\pi]$. 

Use suitable stepsize and annotate the graphics.

````python

from scipy.interpolate import approximate_taylor_polynomial
x = np.linspace(-pi, pi, num=150)
plt.plot(x, np.sin(x), label="sin curve")
for degree in np.arange(1, 7, step=2):
    sin_taylor = approximate_taylor_polynomial(np.sin, 0, degree, 1, order=degree + 2)
    plt.plot(x, sin_taylor(x), label=f"degree={degree}")
plt.axis([-4, 4, -4, 4])
plt.legend(bbox_to_anchor=(1.04, 1))
plt.show()
````
