How to calculate recursive sequence



Fibonacci squence

x[k] = x[k-2] + x[k-1]

x[0] = 1

x[1] = 1

```python
import numpy as np
import matplotlib.pyplot as plt
 
x = np.zeros(8)
x[0] = 1
x[1] = 1
for k in range (2, 8):
  x[k] = x[k-2] + x[k-1]
plt.stem(x)
plt.title('Fibonacci sequence')
plt.show()
```
